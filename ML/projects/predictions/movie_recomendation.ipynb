```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

```

```python
movie_ds = '../data_sets/movies_ds.csv'
movie_data = pd.read_csv(movie_ds)
```

```python
movie_data.head()
```

```python
movie_data.shape
```

```python
movie_data.isnull().sum()
```

```python
selected_features = ["genres", "keywords", 'tagline', 'cast', 'director']
```

```python
# replacing the null values

for feature in selected_features:
    movie_data[feature] = movie_data[feature].fillna("")
```

```python
movie_data.isnull().sum()
```

```python
# combining all the selected features

combined_features = movie_data["genres"]+ movie_data["keywords"]+movie_data["tagline"]+movie_data["cast"]+movie_data["director"]
```

```python
vectorizer = TfidfVectorizer(max_df=1, lowercase=True, stop_words="english")
```

```python
combined_features_vectorizer = vectorizer.fit_transform(combined_features)
```

```python
print(combined_features_vectorizer)
```

```python
# cosine simarity

similarity = cosine_similarity(combined_features_vectorizer)
```

```python
similarity, similarity.shape
```

```python
# getting the movie name from the user

movie_name = input("enter movie name: ")
```

```python
# creating a list with all the movie names given in the data set

list_of_all_titles = movie_data["title"].tolist()
```

```python
# finding the close match for the movie name given by user


find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)
```

```python
find_close_match
```

```python
close_match = find_close_match[0]
```

```python
# finding the index of the movie with title

index_of_the_movie = movie_data[movie_data.title == close_match]['index'].values[0]
```

```python
index_of_the_movie
```

```python
# getting a list of similar score

similarity_score = list(enumerate(similarity[index_of_the_movie]))
```

```python
similarity_score
```

```python
sorted_similarity_movies = sorted(similarity_score, key = lambda x:x[1], reverse=True)
sorted_similarity_movies
```

```python
# printing the similar movies based on the index

print("Movie suggested for you: \n" )

i = 1

for movie in sorted_similarity_movies:
    index = movie[0]
    title_from_index = movie_data[movie_data.index == index]['title'].values[0]
    if (i < 30):
        print(i, '.', title_from_index)
        i+=1
```

```python
movie_name = input(' Enter your favourite movie name : ')

list_of_all_titles = movie_data['title'].tolist()

find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)

close_match = find_close_match[0]

index_of_the_movie = movie_data[movie_data.title == close_match]['index'].values[0]

similarity_score = list(enumerate(similarity[index_of_the_movie]))

sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True) 

print('Movies suggested for you : \n')

i = 1

for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movie_data[movie_data.index==index]['title'].values[0]
  if (i<30):
    print(i, '.',title_from_index)
    i+=1
```
