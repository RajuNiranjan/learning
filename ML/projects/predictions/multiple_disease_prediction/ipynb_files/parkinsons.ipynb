```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
parkinsons_ds = '../../../data_sets/parkinsons.data'

parkinsons_data = pd.read_csv(parkinsons_ds)
```

```python
parkinsons_data.head()
```

```python
parkinsons_data.shape
```

```python
parkinsons_data.isnull().sum()
```

```python
parkinsons_data["status"].value_counts()
```

```python
X = parkinsons_data.drop(columns=['name', 'status'], axis=1)

Y = parkinsons_data['status']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(
    X,
    Y,
    train_size=0.8,
    stratify=Y,
    random_state=2
)
```

```python
X.shape,X_train.shape, X_test.shape
```

```python
model = svm.SVC(kernel="linear")
```

```python
model.fit(X_train, Y_train)
```

```python
X_tarin_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_tarin_prediction, Y_train)

training_data_accuracy
```

```python
X_test_prediction = model.predict(X_train)
test_data_accuracy = accuracy_score(X_test_prediction, Y_train)

test_data_accuracy
```
