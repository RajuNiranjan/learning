## Matplotlib

- usefull for making plots


```python
# importing matplotlib library

import matplotlib.pyplot as plt
```

```python
# importing numpy to get data for out plots

import numpy as np
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
z= np.cos(x)
```

```python
plt.plot(x,y)
plt.show()
```

Print cosine wave


```python
plt.plot(x,z)
plt.show()
```

adding title x-axis and y-axis labels


```python
plt.plot(x,y)
plt.xlabel("Angle")
plt.ylabel("sine vaue")
plt.title("Sine wave")
plt.show()
```

```python
plt.plot(x,z)
plt.xlabel("Angle")
plt.ylabel("cosine vaue")
plt.title("cosine wave")
plt.show()
```

parabola


```python
x = np.linspace(-10, 10, 20)
y = x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x, y, 'g.')
plt.show()
```

```python
plt.plot(x, y, 'rx')
plt.show()
```

```python
x = np.linspace(-5, 5, 50)
plt.plot(x, np.sin(x), 'g-')
plt.plot(x, np.cos(x), 'b--')
plt.show()
```

Bar Plot


```python
fig = plt.figure()
ax= fig.add_axes([0, 0, 1, 1])
languages=["Eng","Tel","Hin","Tamil","Kannada"]
people=[95, 99.9, 39, 89, 10]
ax.bar(languages, people)
plt.xlabel("Languages")
plt.ylabel("people")
plt.title("Language spok by People")
```

Pie Chart


```python
fig1 = plt.figure()
ax= fig1.add_axes([0, 0, 1, 1])
languages=["Eng","Tel","Hin","Tamil","Kannada"]
people=[95, 99.9, 39, 89, 10]
ax.pie(people, labels=languages, autopct="%1.1f%%")
plt.show()
```

Scatte Plot


```python
x = np.linspace(0, 10, 20)
y = np.sign(x)
z = np.cos(x)

fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y, color='g')
ax.scatter(x, z, color='r')
plt.show()
```

3d Scatter Plot


```python
fig3 = plt.figure()
ax=plt.axes(projection='3d')
z=20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x, y, z, c=z, cmap='Blues')
plt.show()
```
