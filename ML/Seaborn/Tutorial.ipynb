## Seaborn:

- Data Visualization Library


```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# setting theme for the plots

sns.set_theme()
```

Note: seaborn having some important built-in datasets


```python
# total bill vs tip dataset

tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.lmplot(
    data=tips,
    x="total_bill",
    y="tip",
    col="time",
    hue='smoker',
    markers=['o','x'],
    height=5
)
```

```python
# load the iris dataset

iris = sns.load_dataset('iris')
```

```python
iris.head()
```

Scatter Plot


```python
sns.scatterplot(
    x='sepal_length',
    y='petal_length',
    hue='species',
    data=iris
)
```

```python
sns.scatterplot(
    x='sepal_length',
    y='petal_width',
    hue='species',
    data=iris
)
```

```python
# loading the titanic dataset

titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

Count Plot


```python
sns.countplot(
    x="class",
    data=titanic
)
```

```python
sns.countplot(
    x="survived",
    data=titanic
)
```

Bar Chart


```python
sns.barplot(
    x="sex",
    y='survived',
    hue="class",
    data=titanic
)
```

```python
# house price data set

from sklearn.datasets import fetch_california_housing
```

```python
house_boston_dataset = fetch_california_housing()
```

```python
house_boston_df = pd.DataFrame(
    house_boston_dataset.data, 
    columns=house_boston_dataset.feature_names
    )
```

```python
house_boston_df["PRICE"]=house_boston_dataset.target
```

```python
house_boston_df.head()
```

Distribution Plot


```python
sns.histplot(
    house_boston_df["PRICE"]
)
```

Correlation:

- Positive Correlation
- Negative Correlation


Heat Map


```python
# constructing Heat Map


correlation = house_boston_df.corr()

plt.figure(figsize=(10,10))
sns.heatmap(
    correlation,
    cbar=True,
    square=True,
    fmt=".2f",
    annot=True,
    # annot_kws=("size":8),
    cmap="Blues"
)
```
