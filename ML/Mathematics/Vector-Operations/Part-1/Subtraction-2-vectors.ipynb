```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
v1 = np.asarray([0, 0, 2, 4])
v2 = np.asarray([0, 0, 4, -5])
sub = v1 - v2
```

```python
sub
```

```python
plt.quiver(v1[0],v1[1],v1[2],v1[3], scale=1, scale_units="xy", angles="xy", color="g")
plt.quiver(v2[0],v2[1],v2[2],v2[3], scale=1, scale_units="xy", angles="xy", color="r")
plt.quiver(sub[0],sub[1],sub[2],sub[3], scale=1, scale_units="xy", angles="xy", color="b")
plt.grid()
plt.xlim(-3, 10)
plt.ylim(-6, 10)
```
