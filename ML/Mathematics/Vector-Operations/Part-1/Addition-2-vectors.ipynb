```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
vector1 = np.asarray([0, 0, 2, 3])

vector2 = np.asarray([0, 0, 3, -2])
```

```python
sum = vector1 + vector2
```

```python
plt.quiver(vector1[0], vector1[1], vector1[2], vector1[3], angles='xy', scale_units='xy', scale=1, color='r')
plt.quiver(sum[0], sum[1], sum[2], sum[3], angles='xy', scale_units='xy', scale=1, color='b')
plt.quiver(vector2[0], vector2[1], vector2[2], vector2[3], angles='xy', scale_units='xy', scale=1, color='g')
plt.xlim(0, 6)
plt.ylim(-3, 6)
plt.grid()
# plt.gca().set_aspect('equal')
```
