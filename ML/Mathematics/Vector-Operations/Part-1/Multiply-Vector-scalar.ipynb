```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
v1 = np.asarray([0, 0, 2, -3])

s = 3
```

```python
mult = s * v1
```

```python
mult
```

```python
plt.quiver(v1[0], v1[1], v1[2], v1[3], scale=1, scale_units='xy', angles='xy', color='r')
plt.quiver(mult[0], mult[1], mult[2], mult[3], scale=1, scale_units='xy', angles='xy', color='b')
plt.xlim(0, 7)
plt.ylim(-10, 1)
```
