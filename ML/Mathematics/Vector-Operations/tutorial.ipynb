Ploting Vector


```python
plt.quiver(0, 0, 4, 5)
plt.show
```

```python
plt.quiver(0, 0, 4, 6, scale_units="xy", angles="xy", scale=1, color="g")
plt.quiver(0, 0, -4, 6, scale_units="xy", angles="xy", scale=1, color="r")
plt.quiver(0, 0, 2, 0, scale_units="xy", angles="xy", scale=1, color="r")
plt.quiver(0, 0, -2, -6, scale_units="xy", angles="xy", scale=1, color="r")
plt.xlim(-8, 8)
plt.ylim(-8, 8)
```
