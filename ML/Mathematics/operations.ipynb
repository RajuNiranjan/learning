### Vector Operations

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme()
```

```python
plt.quiver(0, 0, 5, 4)
plt.show()
```

```python
plt.quiver(0, 0, 5, 4, scale_units='xy', angles='xy', scale=1, color='r')
plt.xlim(-8, 8)
plt.ylim(-8, 8)
plt.show()
```

#### Addition of Two Vectors

```python
vector_1 = np.asarray([0, 0, 2, 3])

vector_2 = np.asarray([0, 0, 3, -2])

sum = vector_1 + vector_2

sum
```

```python
plt.quiver(0, 0, 2, 3, scale_units='xy', angles='xy', scale=1, color='r')
plt.quiver(0, 0, 3, -2, scale_units='xy', angles='xy', scale=1, color='y')
plt.quiver(0, 0, 5, 1, scale_units='xy', angles='xy', scale=1, color='g')
plt.xlim(-8, 8)
plt.ylim(-8, 8)
plt.show()
```

#### Subtraction of Two vectors

```python
vector_1 = np.asarray([0, 0, 2, 3])

vector_2 = np.asarray([0, 0, 3, -2])

sub = vector_1 - vector_2

sub
```

```python
plt.quiver(0, 0, 2, 3, scale_units='xy', angles='xy', scale=1, color='r')
plt.quiver(0, 0, 3, -2, scale_units='xy', angles='xy', scale=1, color='y')
plt.quiver(0, 0, -1, 5, scale_units='xy', angles='xy', scale=1, color='g')
plt.xlim(-8, 8)
plt.ylim(-8, 8)
plt.show()
```

#### Multiplication a vector with a scalar

```python
vector_1 = np.asarray([0, 0, 2, 3])

scalar_value = 2

mult = scalar_value * vector_1

mult
```

```python
plt.quiver(0, 0, 2, 3, scale_units='xy', angles='xy', scale=1, color='r')
plt.quiver(0, 0, 4, 6, scale_units='xy', angles='xy', scale=1, color='b')
plt.xlim(-8, 8)
plt.ylim(-8, 8)
plt.show()
```

```python
vector_1 = np.asarray([0, 0, 2, 3])

scalar_value = -0.5

mult = scalar_value * vector_1

mult
```

```python
plt.quiver(0, 0, 2, 3, scale_units='xy', angles='xy', scale=1, color='r')
plt.quiver(0, 0, -1, -1.5, scale_units='xy', angles='xy', scale=1, color='b')
plt.xlim(-8, 8)
plt.ylim(-8, 8)
plt.show()
```
