```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso
from sklearn import metrics
```

```python
car_ds = pd.read_csv("./cpp.csv")
```

```python
print(f'Fuel {car_ds.fuel.value_counts()} \n\n')
print(f'Seller {car_ds.seller_type.value_counts()} \n\n')
print(f'Transmission {car_ds.transmission.value_counts()} \n\n')
print(f'Owner {car_ds.owner.value_counts()} \n\n')
```

```python
car_ds.replace(
    {
        "fuel":{
            "Petrol":0,
            "Diesel":1,
            "CNG":2,
            "LPG":3,
            "Electric":4
        },
        "seller_type":{
            "Individual":1,
            "Dealer":0,
            "Trustmark Dealer":2
        },
        "transmission":{
            "Manual":0,
            "Automatic":1
        },
        "owner": {
            "First Owner": 0,
            "Second Owner": 1,
            "Third Owner": 2,
            "Fourth & Above Owner": 3,
            "Test Drive Car": 4
        }
        
    },
    inplace=True
)
```

```python
car_ds.head()
```

```python
X = car_ds.drop(columns=["selling_price", "name"], axis=1)
Y = car_ds["selling_price"]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(
    X,
    Y,
    test_size=0.1,
    random_state=2,
)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
train_prediction = model.predict(X_train)
error_score = metrics.r2_score(Y_train, train_prediction)
error_score
```

```python
test_prediction = model.predict(X_test)
error_score = metrics.r2_score(Y_test, test_prediction)
error_score
```

```python
model2= Lasso(alpha=1.0)
```

```python
model2.fit(X_train, Y_train)
```

```python
train_prediction = model2.predict(X_train)
train_accuracy = metrics.r2_score(Y_train,train_prediction)
train_accuracy
```

```python
test_prediction = model2.predict(X_test)
test_accuracy = metrics.r2_score(Y_test,test_prediction)
test_accuracy
```
